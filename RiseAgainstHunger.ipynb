```python
#Finding the top donors based on OwnerId
df2 = pd.DataFrame(op.groupby('OwnerId').sum()['Amount'])
TopOwnID= pd.DataFrame(df2['Amount'].sort_values(ascending=False))
TopOwnID= TopOwnID.add_suffix('OwnerId').reset_index()
TopOwnID = TopOwnID.rename(columns={'AmountOwnerId': 'Amount'})
TopOwnID.head()
```

```python
#Finding the frequent donors
df4= pd.DataFrame(op.groupby('OwnerId').count()['Amount'])
TopOwnIDFreq= pd.DataFrame(df4['Amount'].sort_values(ascending=False))
TopOwnIDFreq= TopOwnIDFreq.add_suffix('OwnerId').reset_index()
TopOwnIDFreq = TopOwnIDFreq.rename(columns={'AmountOwnerId': 'Frequency'})
TopOwnIDFreq.head()
```

```python
#Forming the RFM table
OwnerIdFreqAmount= TopOwnID.merge(TopOwnIDFreq, on='OwnerId', how='inner')
OwnerIdFreqAmount.head()
OwnerRFM= pd.merge(OwnerIdFreqAmount, op[['Deposit_Date__c', 'OwnerId']], on='OwnerId', how='left')
temp_OwnerRFM = OwnerRFM.rename(columns={'Deposit_Date__c': 'Date'})
OwnerRFM= temp_OwnerRFM.groupby('OwnerId').max()
OwnerRFM= OwnerRFM.add_suffix('OwnerId').reset_index()
OwnerRFM = OwnerRFM.rename(columns={'AmountOwnerId': 'Amount', 'FrequencyOwnerId' : 'Frequency', 'DateOwnerId': 'Date'})
OwnerRFM_sorted= OwnerRFM.sort_values(['Date', 'Amount', 'Frequency'], ascending=False)
OwnerRFM_sorted.to_csv("OwnerRFM_sorted.csv")
OwnerRFM_sorted.head()
```

```python
#Sort campaign names based on total amount
df3 = pd.DataFrame(op.groupby('Campaign_Name__c').sum()['Amount'])
Campaign_Name= pd.DataFrame(df3['Amount'].sort_values(ascending=False))
Campaign_Name= Campaign_Name.add_suffix('Campaign_Name').reset_index()
Campaign_Name = Campaign_Name.rename(columns={'Campaign_Name__c': 'Campaign_Name', 'AmountCampaign_Name' : 'Amount'})
TopCampaigns= Campaign_Name.head(20)
TopCampaigns.to_csv("TopCampaigns.csv")
TopCampaigns.head()
```
